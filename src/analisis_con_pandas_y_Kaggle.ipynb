```python
# Identificar duplicados
duplicados = df.duplicated()
# Contar el número de duplicados
num_duplicados = duplicados.sum()
print(f"Número de registros duplicados: {num_duplicados}")
df.head()
```

```python
# Calcular el total después de impuestos y crear una nueva columna
df['Total After Tax'] = df['Total'] - df['Tax 5%']
df.head()
```

```python
# Clasificar ingresos en categorías
conditions = [
    (df['Total After Tax'] < 500),
    (df['Total After Tax'] >= 500) & (df['Total After Tax'] < 1500),
    (df['Total After Tax'] >= 1500)
]
labels = ['Bajo', 'Medio', 'Alto']
df['Income Category'] = pd.cut(df['Total After Tax'], bins=[-float('inf'), 500, 1500, float('inf')], labels=labels)

# Mostrar el DataFrame con las nuevas columnas
print(df.head())
```

```python
#Exploracion de categorias
print('Customer type: ', df['Customer type'].unique())
print('Gender: ', df['Gender'].unique())
print('Product line: ', df['Product line'].unique())
print('Payment: ', df['Payment'].unique())
print('City: ', df['City'].unique())
```

```python
# Agrupaciones de datos utilizando groupby

# Agrupar por 'Product line' y calcular la suma de las ventas
sales_by_product = df.groupby('Product line')['Total'].sum().reset_index()
print("Total Sales by Product Line:")
print(sales_by_product)

# Agrupar por 'Branch' y calcular la suma de las ventas
sales_by_branch = df.groupby('Branch')['Total'].sum().reset_index()
print("\nTotal Sales by Branch:")
print(sales_by_branch)

# Agrupar por 'Product line' y calcular el promedio de las ventas
avg_sales_by_product = df.groupby('Product line')['Total'].mean().reset_index()
print("\nAverage Sales by Product Line:")
print(avg_sales_by_product)
```

```python
# Función personalizada para calcular la proporción
def calculate_sales_proportion(group):
    group['Sales Proportion'] = group['Total'] / group['Total'].sum()
    return group

sales_proportion = df.groupby('Product line').apply(calculate_sales_proportion)
print("\nSales Proportion by Product Line:")
print(sales_proportion)
```
