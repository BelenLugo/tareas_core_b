```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Cargar el dataset desde el archivo CSV local
df=pd.read_csv('../data/tips.csv')

# Crear una figura y un eje
fig, ax = plt.subplots(figsize=(12, 6))

# Gráfico de dispersión con cuadrícula, leyenda, anotaciones y flechas
sns.scatterplot(
    data=df,
    x='total_bill', 
    y='tip', 
    hue='time',  # Diferenciar los puntos por el momento del día (almuerzo o cena)
    style='sex',  # Diferenciar por género
    s=100,        # Tamaño de los puntos
    ax=ax         # Asignar al eje creado
)

# Configuración de títulos y etiquetas
ax.set_title('Relación entre Total Bill y Tip con Anotaciones y Flechas', fontsize=20)
ax.set_xlabel('Total Bill ($)', fontsize=15)
ax.set_ylabel('Tip ($)', fontsize=15)
ax.legend(loc='upper right', fontsize='large', shadow=True, title='Detalles')

# Añadir cuadrícula
ax.grid(True, which='both', linestyle='--', linewidth=0.7, color='gray')

# Añadir anotación
ax.annotate(
    'Mayor tip en la cena', 
    xy=(45, 10),             # Coordenadas del punto anotado
    xytext=(30, 12),         # Posición del texto
    arrowprops=dict(facecolor='black', shrink=0.05)  # Estilo de la flecha
)

# Añadir flecha
ax.arrow(
    10, 8,                   # Coordenadas iniciales
    5, 1,                    # Desplazamiento de la flecha
    head_width=0.5, 
    head_length=0.5, 
    fc='green', ec='green'   # Color de la flecha
)

# Mostrar el gráfico
plt.show()
```
