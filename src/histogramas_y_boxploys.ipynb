```python
columnas_categoricas = ['Country', 'Region','Happiness Rank', 'Happiness Score', 'Standard Error', 'Economy (GDP per Capita)', 'Family', 'Health (Life Expectancy)', 'Freedom', 'Trust (Government Corruption)', 'Generosity', 'Dystopia Residual']
for columna in columnas_categoricas:
    print(f"Valores únicos en '{columna}':")
    print(df[columna].unique())
    print("\n")
```

```python
variables = ['Happiness Score', 'Economy (GDP per Capita)', 'Health (Life Expectancy)', 'Freedom']

# Personalización de colores
colores = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']

# Crear histogramas
plt.figure(figsize=(12, 10))
for i, var in enumerate(variables):
    plt.subplot(2, 2, i+1)
    plt.hist(df[var], bins=15, color=colores[i], edgecolor='black', alpha=0.7)
    plt.title(f'Histograma de {var}')
    plt.xlabel(var)
    plt.ylabel('Frecuencia')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 10))
for i, var in enumerate(variables):
    plt.subplot(2, 2, i+1)
    sns.boxplot(x=df[var], color=colores[i])
    plt.title(f'Boxplot de {var}')
    plt.xlabel(var)
plt.tight_layout()
plt.show()
```

```python
for var, color in zip(variables, colores):
    fig, ax = plt.subplots(1, 2, figsize=(12, 5))

    # Histograma
    ax[0].hist(df[var], bins=15, color=color, edgecolor='black', alpha=0.7)
    ax[0].set_title(f'Histograma de {var}')
    ax[0].set_xlabel(var)
    ax[0].set_ylabel('Frecuencia')

    # Boxplot
    sns.boxplot(x=df[var], ax=ax[1], color=color)
    ax[1].set_title(f'Boxplot de {var}')
    ax[1].set_xlabel(var)

    plt.tight_layout()
    plt.show()
```

Distribución del Puntaje de Felicidad


Histograma: La mayoría de los países tienen un puntaje de felicidad que se concentra entre los valores de [X, Y] (sustituye X e Y con los rangos observados en el histograma). Esto indica que una gran parte de los países tienen niveles moderados de felicidad, con menos países en los extremos.

Boxplot: Se identifican valores atípicos, con algunos países que tienen puntajes de felicidad extremadamente altos o bajos. Estos valores podrían corresponder a naciones con políticas, economías o culturas significativamente diferentes al promedio global.

Distribución del PIB per Cápita

Histograma: La distribución está sesgada hacia la derecha, lo que indica que la mayoría de los países tienen un PIB per cápita bajo, mientras que unos pocos países cuentan con un PIB per cápita excepcionalmente alto. Esto puede reflejar desigualdades económicas globales.

Boxplot: Los valores atípicos se encuentran en la cola derecha, lo que confirma que hay países con una riqueza per cápita mucho mayor al promedio. Es importante estudiar estos valores para entender cómo influye el desarrollo económico en otras variables como la felicidad o la salud.

Distribución de la Expectativa de Vida Saludable

Histograma: La distribución es relativamente simétrica, con un pico alrededor de [Z años] (sustituye Z con el pico observado). Esto sugiere que la expectativa de vida saludable es similar en gran parte de los países.

Boxplot: Se identifican algunos valores atípicos en ambos extremos, mostrando que ciertos países presentan condiciones de salud pública significativamente mejores o peores en comparación con el promedio.
