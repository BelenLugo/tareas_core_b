```python
import pandas as pd
import matplotlib.pyplot as plt
from io import StringIO
import numpy as np

# Datos de ejemplo en formato CSV
data = """
fecha,ventas
2023-01-01,200
2023-01-02,220
2023-01-03,210
2023-01-04,205
2023-01-05,215
2023-01-06,225
"""

# Cargar los datos en un DataFrame de Pandas
df = pd.read_csv(StringIO(data))  # Usar StringIO en lugar de pd.compat.StringIO

# Convertir la columna 'fecha' a formato datetime
df['fecha'] = pd.to_datetime(df['fecha'])  # Corregir el nombre de la columna

# Grfico de lneas de las ventas
plt.figure(figsize=(10, 6))
plt.plot(df['fecha'], df['ventas'])
plt.title("Ventas Diarias")
plt.xlabel("Fecha")
plt.ylabel("Ventas")
plt.grid(True)  # Aadir una cuadrcula para mejor visualizacin
plt.show()

# Crear un grfico de dispersin
plt.figure(figsize=(10, 6))
plt.scatter(df['fecha'], df['ventas'], color='red')
plt.title("Ventas Diarias (Dispersin)")
plt.xlabel("Fecha")
plt.ylabel("Ventas")
plt.grid(True)
plt.show()

# Crear un histograma de las ventas
plt.figure(figsize=(10, 6))
plt.hist(df['ventas'], bins=5, color='green', edgecolor='black')
plt.title("Distribucin de Ventas")
plt.xlabel("Ventas")
plt.ylabel("Frecuencia")
plt.grid(True)
plt.show()
```
